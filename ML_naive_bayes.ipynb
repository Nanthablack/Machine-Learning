```python
#import
import pandas as pd
```

```python
#load data
df = pd.read_csv('Naive Bayes - Sheet1.csv', names = ['income', 'status', 'credit', 'age', 'buy'],index_col = False)
df
```

```python
def calProb(buy,income, status,credit,age):
    buy_prob = len(df[df['buy'] == 1])/len(df)

    income_prob =len(df[(df['income'] == income)&(df['buy'] == 1)])/len(df[df['buy'] == 1])

    status_prob =len(df[(df['status'] == status)&(df['buy'] == 1)])/len(df[df['buy'] == 1])

    credit_prob =len(df[(df['credit'] == credit)&(df['buy'] == 1)])/len(df[df['buy'] == 1])

    age_prob =len(df[(df['age'] == age)&(df['buy'] == 1)])/len(df[df['buy'] == 1])
 
    VNB = buy_prob*income_prob*status_prob*credit_prob*age_prob

    buy_pro = len(df[df['buy'] == 0])/len(df)
  
    income_pro =len(df[(df['income'] == income)&(df['buy'] == 0)])/len(df[df['buy'] == 0])
    
    status_pro =len(df[(df['status'] == status)&(df['buy'] == 0)])/len(df[df['buy'] == 0])
   
    credit_pro =len(df[(df['credit'] == credit)&(df['buy'] == 0)])/len(df[df['buy'] == 0])
    
    age_pro =len(df[(df['age'] == age)&(df['buy'] == 0)])/len(df[df['buy'] == 0])
    
    VNB0 = buy_pro*income_pro*status_pro*credit_pro*age_pro

    if buy == 1:
          resulf = VNB/(VNB0+VNB)
    else:
          resulf = VNB0/(VNB0+VNB)

    return resulf
```

```python
calProb(1,'Medium', 'Studying','Fair','<=30')
```
